```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
statistics = pd.read_csv('statistics.csv')
statistics
```

```python
statistics = statistics.sort_values(by=['time'])
```

```python
ax = statistics.plot(kind='bar', x='algorithm', y='time' )
ax.bar_label(ax.containers[0], labels=[f'{p:.2f}seg ' for p in statistics['time']])


# ax.bar_label(ax.containers[1], labels=[f'{p:.2f} ' for p in statistics['time']], color='white')
plt.xlabel("Algoritmo")
plt.ylabel("Tempo decorrido")
plt.title("Tempo decorrido por algoritmo")


# ax.bar_label(ax.containers[0])
plt.show()
```

```python
ax2 = statistics.plot(kind='bar', x='algorithm', y='nodes_expanded' )
ax2.bar_label(ax.containers[0], labels=[f'{round(p)}' for p in statistics['nodes_expanded']])
ax2.set_ylim([0, 80000])
# ax2.plot(statistics['algorithm'], statistics['time'])
# ax.bar_label(ax.containers[1], labels=[f'{p:.2f} ' for p in statistics['time']], color='white')
plt.xlabel("Algoritmo")
plt.ylabel("Nós expandidos")
plt.title("Nós expandidos por algoritmo")

plt.show()
```

```python
ax3 = statistics.plot(kind='bar', x='algorithm', y='total_path_cost' )
ax3.bar_label(ax.containers[0], labels=[f'{round(p)}' for p in statistics['total_path_cost']])
ax3.set_ylim([0, 1200])
# ax3.plot(statistics['algorithm'], statistics['time'])
# ax.bar_label(ax.containers[1], labels=[f'{p:.2f} ' for p in statistics['time']], color='white')
plt.xlabel("Algoritmo")
plt.ylabel("Custo total do caminho")
plt.title("Custo total por algoritmo")

plt.show()
```

```python
ax4 = statistics.plot(kind='bar', x='algorithm', y='n_steps' )
ax4.bar_label(ax.containers[0], labels=[f'{round(p)}' for p in statistics['n_steps']])
ax4.set_ylim([0, 1200])
# ax3.plot(statistics['algorithm'], statistics['time'])
# ax.bar_label(ax.containers[1], labels=[f'{p:.2f} ' for p in statistics['time']], color='white')
plt.xlabel("Algoritmo")
plt.ylabel("Número de passos")
plt.title("Número de passos por algoritmo")

plt.show()
```
